#### 捕获异常
* 可能引发异常的关键操作放在try子句中，处理异常的代码编写在except子句中
* 如果没有异常发生，则跳过Except的内容，并继续正常流程。如果发生任何异常，它将被Except捕获
* Except可以指定要捕获的异常类型

````python
# import module sys to get the type of exception
import sys
randomList = ['a', 0, 2]
for entry in randomList:
    try:
        print("The entry is", entry)
        r = 1/int(entry)
        break
    except:
        print("Oops!", sys.exc_info()[0], "occurred.")
        print("Next entry.")
        print()
print("The reciprocal of", entry, "is", r)
````

### 3.对象的私有变量和内置方法
* 默认情况下，Python中的成员函数和成员变量都是公开的(public)
* 在python中定义私有变量只需要在变量名或函数名前加上一个（私有）或两个（伪私有）下划线，那么这个函数或变量就是私有/伪私有的了
* 私有变量不可以直接访问，公有变量可以直接访问
* 伪私有变量可以通过**实例.__类名_变量名**格式来强制访问

#### @property装饰器定义只读属性
* 由于python进行属性的定义时，没办法设置私有属性，因此要通过装饰来进行设置。这样可以隐藏属性名，让用户进行使用的时候无法随意修改

````python
class DataSet(object):
    def __init__(self):
        self._images = 10
        self._labels = 1 # 定义属性的名称
    @property
    def images(self): # 方法加入@property后，方法->属性
        return self._images
    @property
    def labels(self):
        return self._labels

A = DataSet()
print(A.images) # 直接调用images，而不可见_images，从而使用户无法修改_images，从而保护了对象的私有变量
````

* 使用```@<attribute_name>.setter```装饰器可以将一个方法转换为可写属性的setter方法

````python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self): # 访问API
        return self._name

    @name.setter
    def name(self, value): # 修改API
        print("called")
        self._name = value

p = Person('Alice')
print(p.name)
p.name = 'Bob'
print(p.name)
````

#### 定义一个迭代器类

````python
class Foo:
    def __init__(self, n):
        self.n = n
    def __iter__(self): # 必须有
        return self
    def __next__(self):
        if self.n >= 8:
            raise StopIteration
        self.n += 1
        return self.n

f1 = Foo(5)
for i in f1: # 循环时调用__iter__和__next__
    print(i)
````

````python
class Infiter:
    step = 2
    def __init__(self, num):
        self.n = num
    def __iter__(self):
        print("called")
        Infiter.step = 3
        return self
    def __next__(self):
        self.n += Infiter.step
        if self.n < 16:
            return self.n
        else:
            raise StopIteration

f2 = Infiter(5)
print(next(f2))
print(next(f2))
print('---------------')
for i in f2:
    print(i)
````

### 5.可调用对象（callable object）
* callable()函数用来判定对象是否能被调用执行
* 普通数据：callable(“hello”) 返回False
* 函数及类定义：callable返回True
    * 类定义可以理解为__call__()方法派生出的所有可执行对象的**公有运行环境(上下文)**
* 普通对象**实例**：callable返回False
    * 实现__call__()方法的对象实例callable返回True
    * 可以实际运行的都是实例，实例处在某个上下文中

````python
class LinePrint:
    def __init__(self, newline = '\n'):
        self.line = 0
        self.rt = newline
    def print(self, x):
        print(self.line, x, end = self.rt)
        self.line += 1

printf = LinePrint()
printf.print("e1")
printf.print("e2")
printf.print("e3")
print(callable(printf)) # 不可执行对象不能直接调用
# printf("ss") # TypeError: 'LinePrint' object is not callable
````

````python
class LinePrint:
    def __init__(self, newline = '\n'):
        self.line = 0
        self.rt = newline
    def __call__(self, x):
        print(self.line, x, end = self.rt)
        self.line += 1
        return x

list(map(LinePrint(), [10, 20, 30])) # 派生一个可执行实例
````

#### 基于类的装饰器
* 基于类的装饰器的实现，必须实现```call```和```init```两个内置函数。```init```：接收被装饰函数```f```
* ```call()``` ：保证类是可调用对象，同时在内部实现对输入函数的装饰逻辑

````python
class Memorize:
    def __init__(self, func):
        self.f = func
        self.memo = {}
    def __call__(self, *args):
        if args[0] not in self.memo:
            self.memo[args[0]] = self.f(*args) # 执行被装饰函数
            print(args[0], 'not in;')
        return self.memo[args[0]]
````

````python
def factorial(k):
    if k < 2:
        return 1
    return k * factorial(k - 1)

factorial1 = Memorize(factorial) # buggy，实例化一个可执行对象，可能出错，要求=左侧必须为factorial（因为factorial是递归函数）
print(factorial1(4))
factorial1(5)
````

````python
factorial = Memorize(factorial) # 要求=左侧必须为factorial（因为factorial是递归函数）
print(factorial(4))
factorial(5)
````

````python
# 更好的写法
@Memorize
def factorial(k):
    if k < 2:
        return 1
    return k * factorial(k - 1)
print(factorial(4))
print(factorial(5))
````

#### 二叉树

````python
class BinaryTree(object):
    def __init__(self, root):
        self.root = root
        self.left = None
        self.right = None
    def insertLeft(self, node):
        if self.left == None:
            self.left = BinaryTree(node)
        else:
            tmp = BinaryTree(node)
            tmp.left = self.left
            self.left = tmp
    def insertRight(self,node):
        if self.right == None:
            self.right = BinaryTree(node)
        else:
            tmp = BinaryTree(node)
            tmp.right = self.right
            self.right = tmp
    def getRightChild(self):
        return self.right
    def getLeftChild(self):
        return self.left
    def setRootVal(self, node):
        self.root = node
    def getRootVal(self):
        return self.root
````

````python
class Person(object): # 基类
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.weight = 'weight'
        
    def talk(self):
        print("A person is talking...")

class Chinese(Person): # 派生类
    def __init__(self, name, age, language): # 先继承，再重构
        # Person.__init__(self, name, age) # 旧式的写法
        super(Chinese, self).__init__(name, age) # 新式的写法，注意参数不是(self, name, age)
        self.language = language

    def talk(self): # 子类重写父类方法
        print('%s is speaking Chinese.' % self.name)
        
    def walk(self):
        print ("A Chinese is walking")

c = Chinese('Xiao Wang', 22, 'Chinese')
c.talk()
c.walk()
````

### 8. 类的静态方法
* 不会被重新创建，直接按名引用
* 静态方法的运行环境恒定（属于类而不属于类的实例），而类方法的运行环境可以随着继承关系而改变

````python
# 实现多个初始化函数
class Book:
    def __init__(self, title):
        self.title = title

    @classmethod
    def class_method_create(cls, title):
        book = cls(title = title)
        return book

    @staticmethod # 也是类方法，在当前类的上下文中，运行环境不会改变
    def static_method_create(title):
        book = Book(title)
        return book

book1 = Book("Use instance_method_creat book instance")
book2 = Book.class_method_create("Use class_method_creat book instance")
book3 = Book.static_method_create("Use static_method_creat book instance")
print(book1.title)
print(book2.title)
print(book3.title)
````

````python
class Foo(object):
    X = 1
    Y = 14

    @staticmethod
    def ave(*mixes): # 父类中的静态方法
        return sum(mixes) / len(mixes)

    @staticmethod
    def static_method():
        print("父类中的静态方法")
        return Foo.ave(Foo.X, Foo.Y) # 静态方法由于不使用相对引用来标定参数，因此不会随着继承到新环境而改变运算逻辑（环境一直为 X = 1, Y = 14）

    @classmethod
    def class_method(cls): # 父类中的类方法
        print("父类中的类方法")
        return cls.ave(cls.X, cls.Y) # 类方法由cls参数自动带入类的环境引用，因此会随着继承到新环境而改变运算逻辑

class Son(Foo):
    X = 3
    Y = 5

print(Son.static_method())
print(Son.class_method())
````

### 四、Python的模块（Modules）
* 是以.py文件组织的实现特定功能的预定义的函数或环境变量代码
* 可以用```import```（ 路径+文件名） 的形式加载到当前代码环境中

````python
%%writefile calc.py

def mod10sum(li):
    return int(sum(li)) % 10

def modXsum(li, x = 10):
    int(x)
    return int(sum(li)) % x
````

````python
import calc as f

w = [1, 2, 3, 4, 5, 6, 7]
print(f.mod10sum(w)) # 要加模块名前缀
print(f.modXsum(w, 8))
````

````python
from calc import mod10sum

print(mod10sum(w)) # 直接用函数名
````

#### ```__name__```属性
* 模块（.py文件）在创建之初会自动加载一些内建变量，```__name__```就是其中之一
* ``` if __name__ == '__main__':```
    * 保护模块私有的执行（调试）代码不被包含到其他模块中
    * 只有当文件被当作脚本运行的时候，```__name__```才会等于```__main__```
